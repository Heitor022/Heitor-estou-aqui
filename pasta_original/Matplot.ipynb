```python
#Matplotlib

#biblicoteca para visualização 2d e 3d, baseada em Numpy
```

```python
import numpy as np
meses = np.array(['jan','fev','mar','abril','maio','junho','julho'])
temperaturas = np.array([22,23,24,25,26,30,35])
len(meses), len(temperaturas)
```

```python
import matplotlib.pyplot as plt
plt
```

```python
plt.figure(figsize=[40,40]) #definição de tamanho do quadro gráfico
```

```python
plt.figure(figsize=[10,10]) #definição de tamanho do quadro gráfico - largura x altura
plt.title('1990')
plt.plot(meses, temperaturas, label = 'temperatura') #visualização do gráfico - "definição de legenda através do label"
plt.yticks([0,10,20,30]) #definição de escala em y
plt.xlabel('meses')
plt.ylabel('temperatura')
plt.legend()
plt.show() #gráfico final
plt.savefig('primeira_aula_pratica_matplot.png')
```

```python
#exercícios Matplot

#para consultar a biblioteca matplot, utilizar da função help(plt.(função)) - exemplo: help(plt.pie)
import matplotlib.pyplot as plt

valores = [15,30,45,10]
rotulos = ['Frogs', 'hogs', 'dogs', 'logs']
explode = [0, .1, 0, 0]
cores = ['blue','orange','green','red']
plt.figure(figsize=(10,10))
plt.pie(valores, 
        explode = explode, 
        labels = rotulos, 
        colors=cores,
#gráfico de pizza (utilizar pie depois da função plt)
#função explode permite separar o pedaço da pizza do inteiro. 
        autopct="%.1f%%",
        pctdistance =.4, 
        shadow = True, #criar sombra
        startangle=90 #rotacionar o gráfico 
        )
plt.show() #exibir o gráfico

```

```python
#exemplo de gráficos com pontos no matplot
import numpy as np #importando numpy para trabalharmos com array e valores fracionados e inteiros dentro deles

x=[.8,2.4,3.3,4,4.4,5]
y=[63,34,22,48,18,5]
tamanho=np.array([10,0,3,0,2,10])*20
cores =['red','green','yellow','yellow','green','red']
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            s=tamanho,
            c=cores)
plt.xticks(np.arange(1.5,5,.5)) #escala e divisão do eixo X 
plt.yticks(np.arange(20,61,10))
plt.show()
```

```python
#outros exemplos

x = np.random.rand(500)
y = np.random.rand(500)
plt.figure()
plt.scatter(x,y,alpha=.6) #alpha é utilizado para a transparência dos pontos
plt.plot()
```

```python
#gráficos de barra

x=['Python','C++','Java','Perl','Scala','Lisp']
y=[10,8,6,4,2,1]

plt.figure(figsize=(7,7))
plt.bar(x,y,
        color="mediumpurple",
        alpha=.8,
        edgecolor='black',
        linewidth=.9) #silhueta das barras
plt.ylabel('Usage')
plt.xlabel('Programming Language Usagel')
plt.ylim(0,10) #range da escala
plt.show()
```

```python
#GRÁFICO EM BARRA INVERTIDA

x=['Python','C++','Java','Perl','Scala','Lisp']
y=[10,8,6,4,2,1]

plt.figure(figsize=(7,7))
plt.barh(x,y,
        color="mediumpurple",
        alpha=.8,
        edgecolor='black',
        linewidth=.9) #silhueta das barras
plt.ylabel('Usage')
plt.xlabel('Programming Language Usagel')
plt.ylim(0,10) #range da escala
plt.show()
```
